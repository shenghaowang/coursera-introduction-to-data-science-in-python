```python
import sys
import pymysql as pymysql
import xmltodict as xmltodict
```

```python
def get_all_question_to_box_map(task_ids):
   all_question_to_box_map = {}
   for task_id in task_ids:
       connection = pymysql.connect(host='annotation.cz7savcjxymi.ap-southeast-1.rds.amazonaws.com',
                                    user='annotation_rw',
                                    password='9DqPrEJGCEXFf6pP',
                                    db='rec_annotation',
                                    charset='utf8mb4',
                                    cursorclass=pymysql.cursors.DictCursor)
       with connection.cursor() as cursor:
           sql = 'select content from vs_question where task_id=' + str(task_id)
           cursor.execute(sql)
           ret = cursor.fetchall()

       
       for line in ret:
           content = xmltodict.parse(line['content'])
           question = content['QuestionForm']['Question']
           #print(question['QuestionContent'])
           qid = question['QuestionIdentifier']
           #print(qid)
           boxes = question['QuestionContent']['ResultBoxes']
           #print(boxes)
           all_question_to_box_map[qid] = boxes

   return all_question_to_box_map
```
